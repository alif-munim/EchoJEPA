```python
VIEW_PROB_COL = {v: v for v in CANONICAL_VIEWS}  # p_* columns already match
# URI_COL = "mp4_uri" if "mp4_uri" in df.columns else "URI"
URI_COL = "URI"
STUDY_COL = "DeidentifiedStudyID"
LABEL_COL = "Value"  # text label column
MISSING_TOKEN = "MISS"  # will be written in CSV for missing views
```

```python
label_mapping = {"normal": 0, "reduced": 1}
df["label_int"] = df[LABEL_COL].map(label_mapping)
if df["label_int"].isna().any():
    raise ValueError(f"Unmapped labels: {df.loc[df['label_int'].isna(), LABEL_COL].unique().tolist()}")
```

```python
# ----- per-study selection: take highest-confidence URI per view -----
def select_views_for_study(g: pd.DataFrame) -> pd.Series:
    out_paths = {}
    out_confs = {}
    for v in CANONICAL_VIEWS:
        pcol = VIEW_PROB_COL[v]
        # robust: ignore rows with missing/empty URI
        g_valid = g[g[URI_COL].astype(str).str.len() > 0].copy()
        if g_valid.empty or (pcol not in g_valid.columns):
            out_paths[v] = MISSING_TOKEN
            out_confs[v] = np.nan
            continue
        idx = g_valid[pcol].astype(float).idxmax()
        best_p = float(g_valid.loc[idx, pcol])
        best_uri = str(g_valid.loc[idx, URI_COL]) if best_p > 0 else MISSING_TOKEN
        out_paths[v] = best_uri
        out_confs[v] = best_p if best_uri != MISSING_TOKEN else np.nan

    # label: majority vote (mode) over the group
    label_mode = g["label_int"].mode()
    label_int = int(label_mode.iloc[0])

    # row that will be written to CSV (space-separated URIs in canonical order + label)
    csv_row = " ".join(out_paths[v] for v in CANONICAL_VIEWS) + f" {label_int}"

    # pack: one column per view path + label
    rec = {f"{v}": out_paths[v] for v in CANONICAL_VIEWS}
    rec["label"] = label_int
    rec["CSV_Row"] = csv_row
    # also include confidences for preview
    for v in CANONICAL_VIEWS:
        rec[f"conf_{v}"] = out_confs[v]
    return pd.Series(rec)
```

```python
per_study = (
    df.groupby(STUDY_COL, sort=False, as_index=True)
      .apply(select_views_for_study)
)

# ----- preview table you asked for -----
preview_cols = [STUDY_COL, "CSV_Row"] + CANONICAL_VIEWS  # show chosen URIs per slot
conf_cols = [f"conf_{v}" for v in CANONICAL_VIEWS]
preview = per_study.reset_index()[[STUDY_COL, "CSV_Row", *conf_cols]]
print(preview.head())
```

```python
import numpy as np

N_VIEWS = 9
MISS = "MISS"

def enforce_at_least_one_miss_headerless(df, n_views=N_VIEWS, miss_token=MISS, seed=123):
    """Ensure each CSV_Row has at least one MISS among the n_views paths."""
    rng = np.random.default_rng(seed)
    df = df.copy()

    new_rows = []
    for s in df["CSV_Row"].astype(str).tolist():
        parts = s.strip().split()
        if len(parts) < n_views + 1:
            raise ValueError(f"CSV_Row has {len(parts)} fields; expected {n_views}+label: {s[:200]}")

        paths = parts[:n_views]
        label = parts[-1]

        if all(p != miss_token for p in paths):
            j = int(rng.integers(0, n_views))
            paths[j] = miss_token

        new_rows.append(" ".join(paths + [label]))

    df["CSV_Row"] = new_rows
    return df

train_df_miss = enforce_at_least_one_miss_headerless(train_df, n_views=N_VIEWS)
val_df_miss   = enforce_at_least_one_miss_headerless(val_df,   n_views=N_VIEWS)
test_df_miss  = enforce_at_least_one_miss_headerless(test_df,  n_views=N_VIEWS)

# write headerless (paths + label) with MISS_ prefix
train_df_miss["CSV_Row"].to_csv(f"csv/MISS_multi9_{label_names}_train.csv", index=False, header=False)
val_df_miss["CSV_Row"].to_csv(  f"csv/MISS_multi9_{label_names}_val.csv",   index=False, header=False)
test_df_miss["CSV_Row"].to_csv( f"csv/MISS_multi9_{label_names}_test.csv",  index=False, header=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# --- config -------------------------------------------------------------------
# label_names = "mv_function_a4c"        # <-- set your base filename
N = 8                                   # number of URIs per study to emit
test_size = 0.20
val_size_within_trainval = 0.20
RNG = 42

# --- load & label map ---------------------------------------------------------
df = all_rvfx_labels
# df = pd.read_csv(f"csv/{label_names}.csv")

label_mapping = {
    "normal": 0,
    "reduced": 1,
}

df["Value_numeric"] = df["Value"].map(label_mapping)

# sanity
if df["Value_numeric"].isna().any():
    bad = df.loc[df["Value_numeric"].isna(), "Value"].unique().tolist()
    raise ValueError(f"Unmapped labels present: {bad}")

# --- collapse rows to one row per DeidentifiedStudyID with N paths ------------
def rows_for_study(g: pd.DataFrame, n: int) -> pd.Series:
    # keep original order as in the CSV
    uris = g["URI"].tolist()
    if len(uris) == 0:
        # should not happen, but guard anyway
        uris = [""] * n
    elif len(uris) < n:
        uris = uris + [uris[-1]] * (n - len(uris))  # duplicate last until N
    else:
        uris = uris[:n]

    # label: take the mode; if tie, .mode() returns sorted values -> take first
    label_mode = g["Value_numeric"].mode()
    label = int(label_mode.iloc[0])

    out = {f"path{i+1}": uris[i] for i in range(n)}
    out["label"] = label
    return pd.Series(out)
```

```python
# groupby preserves intra-group row order by default
study_df = (
    df.groupby("DeidentifiedStudyID", sort=False)
      .apply(rows_for_study, n=N)
      .reset_index(drop=True)
)

# --- split at the STUDY level (to avoid leakage) ------------------------------
train_val_df, test_df = train_test_split(
    study_df,
    test_size=test_size,
    stratify=study_df["label"],
    random_state=RNG,
)

train_df, val_df = train_test_split(
    train_val_df,
    test_size=val_size_within_trainval,
    stratify=train_val_df["label"],
    random_state=RNG,
)

# --- stats --------------------------------------------------------------------
def pct(n, d): return 100.0 * n / max(1, d)

print(f"Total studies: {len(study_df)}")
print(f"Train: {len(train_df)} ({pct(len(train_df), len(study_df)):.1f}%)")
print(f"Val:   {len(val_df)} ({pct(len(val_df), len(study_df)):.1f}%)")
print(f"Test:  {len(test_df)} ({pct(len(test_df), len(study_df)):.1f}%)")
```

```python
import pandas as pd  
from sklearn.model_selection import train_test_split  
  
# Read mitral valve regurgitation labels for A4C videos
df = pd.read_csv(f"{label_names}.csv")  
  
# Create label mapping for severity levels  
label_mapping = {  
    'normal': 0,  
    'reduced': 1
}  

def build_train_test(df, label_mapping):  
    # Convert string labels to integers  
    df['Value_numeric'] = df['Value'].map(label_mapping)  
      
    # Verify all labels were mapped correctly  
    if df['Value_numeric'].isna().any():  
        print("Warning: Some labels couldn't be mapped!")  
        print("Unmapped labels:", df[df['Value_numeric'].isna()]['Value'].unique())  
      
    # First split: separate out test set (60% train+val, 40% test)  
    train_val_df, test_df = train_test_split(  
        df,   
        test_size=0.2,  # 20% for test  
        stratify=df['Value_numeric'],   
        random_state=42  
    )  
      
    # Second split: split train+val into train and validation (64% train, 16% val, 20% test)  
    train_df, val_df = train_test_split(  
        train_val_df,  
        test_size=0.2,  # 20% of remaining 80% = 16% of total  
        stratify=train_val_df['Value_numeric'],  
        random_state=42  
    )  

    return train_df, val_df, test_df
```

# Make a Tiny Subset

```python
import numpy as np
import pandas as pd

def sample_per_class(df: pd.DataFrame,
                     label_col: str = "Value_numeric",
                     n_per_class: int = 500,
                     seed: int = 0) -> pd.DataFrame:
    """Balanced down-sample for fast overfit/debug runs."""
    rng = np.random.default_rng(seed)
    return (
        df.groupby(label_col, group_keys=False)
          .apply(lambda g: g.sample(min(len(g), n_per_class), random_state=rng.integers(1e9)))
          .reset_index(drop=True)
    )

# --- choose sizes ----------------------------------------------------
train_small = sample_per_class(train_df, n_per_class=1000)   # 2 000 rows (1 000 × 2 classes)
val_small   = sample_per_class(val_df,   n_per_class=250)    #   500 rows
test_small  = sample_per_class(test_df,  n_per_class=250)    #   500 rows

# --- save in V-JEPA 2 format ----------------------------------------
for name, df_small in [("train", train_small),
                       ("val",   val_small),
                       ("test",  test_small)]:
    df_small[["URI", "Value_numeric"]].to_csv(
        f"{label_names}_{name}_tiny.csv",
        header=False, index=False, sep=" "
    )

# quick sanity check
for split, d in [("train", train_small), ("val", val_small), ("test", test_small)]:
    counts = d["Value_numeric"].value_counts().sort_index().to_dict()
    print(split, len(d), counts)
```

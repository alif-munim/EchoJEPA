```python
import pandas as pd

# ---------------------------------------------------------
# 1. Prepare the Video Metadata (The 18M Dataset)
# ---------------------------------------------------------

# Merge your view and color dataframes (assuming they align or merging on s3_uri)
# If they are already aligned by index, you can just concat. 
# Ideally, merge on s3_uri to be safe:
df_videos = pd.merge(df_view, df_color[['s3_uri', 'is_color']], on='s3_uri', how='inner')

# FILTER: Keep only Color videos and the two views we care about
df_relevant = df_videos[
    (df_videos['is_color'] == 'Yes') & 
    (df_videos['prediction'].isin(['A4C', 'PSAX-AV']))
].copy()
```

```python
# ---------------------------------------------------------
# 2. Extract Matching Key (Study UID)
# ---------------------------------------------------------

# We need to extract the Study UID from the s3_uri to match 'uhn_df'.
# Looking at your screenshot: 
# s3_uri: ".../echo-study/1.2.276.0.7230010.../..."
# uhn_df['DeidentifiedStudyID']: "1.2.276.0.7230010..."

# This function splits the path and grabs the folder name which is the Study UID
def extract_study_id(uri):
    # Split by '/' and find the part after 'echo-study'
    parts = uri.split('/')
    try:
        # Assuming the structure is .../echo-study/<StudyUID>/<VideoFile>
        # Adjust index if your path structure is slightly different
        study_idx = parts.index('echo-study') + 1 
        return parts[study_idx]
    except (ValueError, IndexError):
        return None

df_relevant['match_id'] = df_relevant['s3_uri'].apply(extract_study_id)
```

```python
# ---------------------------------------------------------
# 3. Find Highest Confidence Video per View per Study
# ---------------------------------------------------------

# Sort by Study, View, and Confidence (High to Low)
df_sorted = df_relevant.sort_values(by=['match_id', 'prediction', 'confidence'], ascending=[True, True, False])

# Group by Study and View, taking the first one (which is now the highest confidence)
df_best = df_sorted.groupby(['match_id', 'prediction']).head(1)
```

```python
import pandas as pd

# ---------------------------------------------------------
# 1. The ROBUST ID Extractor
# ---------------------------------------------------------
def extract_study_id_robust(uri):
    # Split the path into parts
    parts = uri.split('/')
    
    # Iterate through parts to find the Study UID
    # We look for the part starting with '1.2.276' (UHN OID root)
    # This ignores whether it's in 'echo-study', 'echo-study-2', or anywhere else.
    for part in parts:
        if part.startswith('1.2.276'):
            return part
            
    return None

# ---------------------------------------------------------
# 2. Re-Run the Extraction & Matching
# ---------------------------------------------------------

# Apply the new extractor
df_view['match_id'] = df_view['s3_uri'].apply(extract_study_id_robust)

# Merge video data with color data
df_videos = pd.merge(df_view, df_color[['s3_uri', 'is_color']], on='s3_uri', how='inner')

# Filter for the relevant videos (Color + Correct Views)
df_relevant = df_videos[
    (df_videos['is_color'] == 'Yes') & 
    (df_videos['prediction'].isin(['A4C', 'PSAX-AV']))
].copy()

# Sort by Confidence (High -> Low) so we pick the best one
df_sorted = df_relevant.sort_values(by=['match_id', 'prediction', 'confidence'], ascending=[True, True, False])

# Keep only the single best video per view, per study
df_best = df_sorted.groupby(['match_id', 'prediction']).head(1)

# Pivot to Wide Format (One row per study)
df_pivot = df_best.pivot(index='match_id', columns='prediction', values='s3_uri').reset_index()
df_pivot.columns.name = None
df_pivot = df_pivot.rename(columns={'A4C': 'path_A4C', 'PSAX-AV': 'path_PSAX_AV'})
```

```python
# Final Merge with Labels
final_df = pd.merge(
    uhn_df[['DeidentifiedStudyID', 'Value', 'PATIENT_ID', 'STUDY_DATE', 's3_key']], 
    df_pivot,
    left_on='DeidentifiedStudyID',
    right_on='match_id',
    how='inner'
)

# Clean up
final_df = final_df[['s3_key', 'DeidentifiedStudyID', 'PATIENT_ID', 'STUDY_DATE', 'path_A4C', 'path_PSAX_AV', 'Value']]

print(f"Original Match Count: ~24,132")
print(f"New Match Count:      {len(final_df)}")
print(final_df.head())
```

# Create Splits

```python
import pandas as pd
import numpy as np

def create_patient_disjoint_temporal_split(df, 
                                           patient_col='PATIENT_ID', 
                                           time_col='STUDY_DATE', 
                                           ratios=[0.8, 0.1, 0.1]):
    """
    Splits dataframe strictly by patient, ordered by time, respecting data volume ratios.
    """
    # 1. Get unique patients with their temporal metadata and study volume
    patient_meta = df.groupby(patient_col).agg(
        last_study_date=(time_col, 'max'),
        study_count=('s3_key', 'count') # Or any non-null col
    ).reset_index()

    # 2. Sort patients by time (Temporal aspect)
    patient_meta = patient_meta.sort_values(by='last_study_date')

    # 3. Calculate cumulative counts to find split indices based on VOLUME (not just patient count)
    total_studies = len(df)
    patient_meta['cumulative_studies'] = patient_meta['study_count'].cumsum()
    
    # Define thresholds
    train_cutoff = total_studies * ratios[0]
    val_cutoff = total_studies * (ratios[0] + ratios[1])

    # 4. Assign split labels
    def get_split_label(cumulative_count):
        if cumulative_count <= train_cutoff:
            return 'train'
        elif cumulative_count <= val_cutoff:
            return 'val'
        else:
            return 'test'

    patient_meta['split'] = patient_meta['cumulative_studies'].apply(get_split_label)

    # 5. Merge split labels back to original dataframe
    df_split = df.merge(patient_meta[[patient_col, 'split']], on=patient_col, how='left')

    # Create the subsets
    train_df = df_split[df_split['split'] == 'train'].copy()
    val_df = df_split[df_split['split'] == 'val'].copy()
    test_df = df_split[df_split['split'] == 'test'].copy()

    return train_df, val_df, test_df

# --- Usage ---

# Ensure STUDY_DATE is datetime (looks like you did this in Image 3, but good to double check)
final_df['STUDY_DATE'] = pd.to_datetime(final_df['STUDY_DATE'])

train_df, val_df, test_df = create_patient_disjoint_temporal_split(final_df)

# --- Verification & Stats ---
print(f"Total Studies: {len(syn_df_labels)}")
print(f"Train: {len(train_df)} ({len(train_df)/len(uhn_df):.1%}) | Dates: {train_df['STUDY_DATE'].min().date()} to {train_df['STUDY_DATE'].max().date()}")
print(f"Val:   {len(val_df)} ({len(val_df)/len(uhn_df):.1%}) | Dates: {val_df['STUDY_DATE'].min().date()} to {val_df['STUDY_DATE'].max().date()}")
print(f"Test:  {len(test_df)} ({len(test_df)/len(uhn_df):.1%}) | Dates: {test_df['STUDY_DATE'].min().date()} to {test_df['STUDY_DATE'].max().date()}")

# Verify Patient Disjointness
train_pats = set(train_df['PATIENT_ID'])
val_pats = set(val_df['PATIENT_ID'])
test_pats = set(test_df['PATIENT_ID'])

assert len(train_pats.intersection(val_pats)) == 0, "Leakage between Train and Val!"
assert len(train_pats.intersection(test_pats)) == 0, "Leakage between Train and Test!"
assert len(val_pats.intersection(test_pats)) == 0,   "Leakage between Val and Test!"

print("\nSUCCESS: Splits are patient-disjoint and temporally ordered.")
```

# Statistics

```python
# 1. Clean the data (RVSP is usually 15-140 mmHg)
def clean_rvsp(df):
    initial_len = len(df)
    
    # RVSP PHYSIOLOGY BOUNDS
    # Min: 10 (Below this is likely an error or Mean Pressure)
    # Max: 170 (Eisenmenger syndrome/Severe PH can go >120. 100 is too low a cutoff.)
    df_clean = df[(df['Value'] >= 10) & (df['Value'] <= 170)].copy()
    
    dropped_count = initial_len - len(df_clean)
    if dropped_count > 0:
        print(f"Dropped {dropped_count} rows with invalid RVSP values (keeping 10-170 mmHg).")
        
    return df_clean
```

```python
# 1. Create the simplified DataFrames
cols_to_keep = ['path_A4C', 'path_PSAX_AV', 'Value']

train_simple = train_df[cols_to_keep].copy()
val_simple = val_df[cols_to_keep].copy()
test_simple = test_df[cols_to_keep].copy()

# Apply the basic cleaning (removing negatives/impossible values)
train_clean = clean_rvsp(train_simple)
val_clean = clean_rvsp(val_simple)
test_clean = clean_rvsp(test_simple)

# 3. Function to print detailed statistics
def print_split_stats(name, df):
    print(f"================ {name} Set ================")
    print(f"Total Studies: {len(df)}")
    
    # Check for empty paths or null values
    missing_count = df.isnull().sum().sum()
    if missing_count > 0:
        print(f"WARNING: Found {missing_count} missing values!")
    
    # Calculate Skewness
    # RVSP is usually RIGHT skewed (Positive skew):
    # Most people are Normal (20-30), with a long tail of high pressure (PH).
    skewness = df['Value'].skew()
    
    stats = df['Value'].describe(percentiles=[0.05, 0.25, 0.5, 0.75, 0.95])
    
    print("\nTarget ('RVSP in mmHg') Statistics:")
    print(f"  Count: {int(stats['count'])}")
    print(f"  Mean:  {stats['mean']:.2f} mmHg")
    print(f"  Std:   {stats['std']:.2f}")
    print(f"  Skew:  {skewness:.2f} (expect > 0 for medical data)")
    print("-" * 20)
    print(f"  Min:   {stats['min']:.2f}")
    print(f"   5%:   {stats['5%']:.2f}  <-- Normal physiological low")
    print(f"  25%:   {stats['25%']:.2f}")
    print(f"  Median:{stats['50%']:.2f}  <-- Typical Healthy/Mild (~25-30)")
    print(f"  75%:   {stats['75%']:.2f}  <-- Likely Pulmonary Hypertension starts here")
    print(f"  95%:   {stats['95%']:.2f}  <-- Severe Pulmonary Hypertension")
    print(f"  Max:   {stats['max']:.2f}")
    print("\n")

# 4. Run the stats
print_split_stats("TRAIN", train_clean)
print_split_stats("VALIDATION", val_clean)
print_split_stats("TEST", test_clean)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import pickle

# 1. Initialize the Scaler
scaler = StandardScaler()

# 2. Fit ONLY on the Training Set
# This learns the Mean (~35) and Std (~14) of your training data
train_values = train_clean['Value'].values.reshape(-1, 1)
scaler.fit(train_values)

print(f"RVSP Scaler fitted. Mean: {scaler.mean_[0]:.4f} mmHg, Std: {scaler.scale_[0]:.4f}")

# 3. Transform All Splits
# Creates 'norm_value' (e.g., 35mmHg -> 0.0, 65mmHg -> +2.1)
train_clean['norm_value'] = scaler.transform(train_clean['Value'].values.reshape(-1, 1))
val_clean['norm_value']   = scaler.transform(val_clean['Value'].values.reshape(-1, 1))
test_clean['norm_value']  = scaler.transform(test_clean['Value'].values.reshape(-1, 1))

# 4. Save the Scaler
# CRITICAL: Save as 'rvsp_scaler.pkl' so you don't overwrite your LVEF scaler
with open('rvsp_scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

# --- Verification ---
print("\nNormalization Check (Train Set):")
print(train_clean['norm_value'].describe())

print("\nExample Data (showing original mmHg and normalized Z-score):")
print(train_clean[['Value', 'norm_value']].head(3))
```

```python
import os
import pandas as pd

# 1. Define your output directory again
output_dir = 'csv'  # Or 'csv' depending on where you want them
os.makedirs(output_dir, exist_ok=True)

def write_strict_multiview_csv(df, filename):
    # 1. Drop rows where ANY of the required columns are NaN/None
    # This ensures every row has [A4C, PSAX, Label]
    df_clean = df.dropna(subset=['path_A4C', 'path_PSAX_AV', 'norm_value']).copy()
    
    dropped = len(df) - len(df_clean)
    if dropped > 0:
        print(f"⚠️  Dropped {dropped} incomplete rows from {filename} (missing one view).")
    
    output_path = os.path.join(output_dir, filename)
    
    # 2. Select columns
    export_df = df_clean[['path_A4C', 'path_PSAX_AV', 'norm_value']]
    
    # 3. Write to CSV with space delimiter
    export_df.to_csv(
        output_path, 
        sep=' ',       
        header=False,  
        index=False    
    )
    
    print(f"✅ Saved {len(export_df)} complete rows to: {output_path}")

# 2. Regenerate the CSVs
# (Assuming train_clean, val_clean, test_clean are still in memory. 
#  If not, reload them from your previous steps before running this.)

print("Regenerating cleaned CSVs...")
write_strict_multiview_csv(train_clean, 'rvsp_train.csv')
write_strict_multiview_csv(val_clean,   'rvsp_val.csv')
write_strict_multiview_csv(test_clean,  'rvsp_test.csv')
```

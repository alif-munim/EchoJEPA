# Mitral Valve

```python
# Filter for rows containing 'pacemaker' in HLCODE
hl_mvr = finding_intersects[finding_intersects['HLCODE'].str.contains('mitral valve', case=False, na=False)]
```

```python
hl_mvr.head()
```

```python
codes = finding_intersects['HLCODE'].str.strip().str.lower()
hl_mvr = finding_intersects[codes.str.contains('mitral valve.*regurgitation', na=False)]
print(len(hl_mvr))
hl_mvr.head()
```

```python
keep = [
    'local^mitral valve^function^regurgitation^trivial^358',
    'mitral valve^function^regurgitation^mild',
    'mitral valve^function^regurgitation^mild to moderate',
    'mitral valve^function^regurgitation^no evidence of',
    'mitral valve^function^regurgitation^moderate',
    'mitral valve^function^regurgitation^severe',
    'mitral valve^function^regurgitation^moderate to severe'
]

hl_mvr_sel = hl_mvr[hl_mvr['HLCODE'].isin(keep)].copy()
```

```python
mapping_two = {
    'mitral valve^function^regurgitation^no evidence of': 'trivial',
    'local^mitral valve^function^regurgitation^trivial^358': 'trivial',
    'mitral valve^function^regurgitation^mild': 'trivial',
    'mitral valve^function^regurgitation^mild to moderate': 'moderate_or_greater',
    'mitral valve^function^regurgitation^moderate': 'moderate_or_greater',
    'mitral valve^function^regurgitation^moderate to severe': 'moderate_or_greater',
    'mitral valve^function^regurgitation^severe': 'moderate_or_greater'
}

hl_mvr_sel['Regurgitation_Binary'] = hl_mvr_sel['HLCODE'].map(mapping_two)
```

```python
print(hl_mvr_sel['Regurgitation_Binary'].dropna().value_counts())
```

```python
# REP_IDs that appear in the pacemaker dataframe
hl_mvr_ids = hl_mvr_sel['REP_ID'].dropna().unique()
print(len(hl_mvr_ids))

# Split hl_link into pacemaker / no-pacemaker subsets
hl_mvr_labels = hl_link[hl_link['REP_ID'].isin(hl_mvr_ids)].copy()
```

```python
hl_mvr_labels_final = hl_mvr_sel.merge(
    hl_mvr_labels,
    on='REP_ID',
    how='inner',
    suffixes=('', '_drop')  # keep REP_ID as-is, add suffix to duplicates
)

# Remove the duplicate REP_ID column if created
if 'REP_ID_drop' in hl_mvr_labels_final.columns:
    hl_mvr_labels_final = hl_mvr_labels_final.drop(columns=['REP_ID_drop'])

print(hl_mvr_labels_final.columns)
print(hl_mvr_labels_final.shape)
display(hl_mvr_labels_final.head())
```

```python
hl_mvr_labels_final = hl_mvr_labels_final.merge(
    aws_heartlab[['OriginalStudyID', 's3_key']],  # only needed columns
    on='OriginalStudyID',
    how='left'
)

print(hl_mvr_labels_final.shape)
display(hl_mvr_labels_final.head())
```

```python
hl_mvr_labels_final = hl_mvr_labels_final.dropna(subset=['s3_key'])

print(hl_mvr_labels_final.shape)
display(hl_mvr_labels_final.head())
```

```python
# Create a boolean mask to identify matching StudyRef values
matching_mask = syn_mv_obs['StudyRef'].isin(aws_uhn['STUDY_REF'])

# Count the number of matches (True values)
number_of_matches = matching_mask.sum()

# Get the total number of unique studies in mv_obs for context
total_unique_studies = syn_mv_obs['StudyRef'].nunique()

print(f"\nTotal unique StudyRef values in mv_obs: {total_unique_studies}")
print(f"Number of 'StudyRef' values from mv_obs that are in aws_uhn: {number_of_matches}")
print(f"Percentage of matching studies: {(number_of_matches / total_unique_studies) * 100:.2f}%")
```

```python
import pandas as pd

# Assuming 'mv_obs' and 'aws_uhn' are your DataFrames loaded as in the screenshots.

# Step 1: Merge the two DataFrames
# We'll merge 'mv_obs' with 'aws_uhn' using 'StudyRef' and 'STUDY_REF' as the keys.
merged_df = pd.merge(syn_mv_obs, aws_uhn, left_on='StudyRef', right_on='STUDY_REF', how='inner')

# Step 2: Define the desired columns in the correct order
final_columns = [
    'STUDY_REF',
    's3_key',
    'Value',
    'Regurgitation_Binary',
    'STUDY_DATE',
    'STUDY_TIME',
    'DeidentifiedStudyID',
    'OriginalStudyID'
]

# # Step 3: Create the new table with only the specified columns
syn_mvr_labels = merged_df[final_columns]

# replacement_map = {
#     'none': 'trace',
#     'moderate_to_severe': 'severe'
# }

# # Apply the replacement to the 'Value' column
# mvr_labels['Value'] = mvr_labels['Value'].replace(replacement_map)
```

```python
# Columns you want to keep
keep_cols = [
    'STUDY_REF','REP_ID','HLCODE', 'Value', 'Regurgitation_Binary','s3_key',  'STUDY_DATE',
    'STUDY_TIME', 'DeidentifiedStudyID', 'OriginalStudyID',
     
    'OriginalPatientID', 'DeidentifiedPatientID', 'SERI_ID', 'STUDY_ID',
    'STUDY_INSTANCE_UID'
]

# Combine both dataframes
uhn_mvr = pd.concat([syn_mvr_labels, hl_mvr_labels_final], ignore_index=True, sort=False)

# Drop overlapping OriginalStudyIDs (keep first occurrence)
uhn_mvr = uhn_mvr.drop_duplicates(subset='OriginalStudyID', keep='first')

# Restrict to selected columns (any missing columns will be added as NaN)
uhn_mvr = uhn_mvr.reindex(columns=keep_cols)

print(uhn_mvr.shape)
display(uhn_mvr.head())
```
